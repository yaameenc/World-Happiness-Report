```python
# Import necessary libraries, load the dataset, and fit a multiple linear regression model to estimate
# the happiness score based on various predictor variables such as GDP per capita, social support, etc.
import pandas as pd
import statsmodels.api as sm
from io import BytesIO

# Load the dataset
data = pd.read_csv(BytesIO(uploaded['2019.csv']))

# Select predictors and dependent variable
predictors = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
dependent_variable = 'Score'

# Prepare the data by extracting predictor and dependent variable columns
X = data[predictors]
y = data[dependent_variable]

# Add a constant for the regression model to account for the intercept term
X = sm.add_constant(X)

# Fit the multiple regression model using ordinary least squares method
model = sm.OLS(y, X).fit()

# Print the summary of the regression model, which includes coefficients, t-statistics, and R-squared values
print(model.summary())
```

```python
# Check for missing values in each column of the dataset to ensure data integrity
print("Number of missing values in each column:")
print(data.isna().sum())
```

```python
# Identify outliers in the dataset using the interquartile range (IQR) method and remove them for a more robust regression model
Q1 = data.quantile(0.25)
Q3 = data.quantile(0.75)
IQR = Q3 - Q1
data_outliers = data[((data < (Q1 - 1.5 * IQR)) | (data > (Q3 + 1.5 * IQR))).any(axis=1)]

# Remove the outliers from the dataset
data_no_outliers = data.drop(data_outliers.index)

# Define the predictors without 'Overall rank'
predictors_without_rank = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']

# Calculate the correlation matrix
corr_matrix = data_no_outliers[predictors_without_rank].corr()

# Print the correlation matrix as a table
print("Correlation Matrix of Predictor Variables:")
print(corr_matrix.to_string(line_width=1000))

# Display the correlation matrix using a heatmap
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Correlation Matrix of Predictor Variables')
plt.show()
```

The heatmap above shows the correlation between predictor variables. It helps us identify if there is multicollinearity, which occurs when two or more predictor variables are highly correlated. Multicollinearity can make it challenging to interpret the coefficients in a multiple regression model and can lead to incorrect conclusions about the relationships between predictor variables and the dependent variable. By examining the heatmap, we can identify potential multicollinearity issues and decide how to handle them before fitting the multiple regression model.

```python
# Evaluate the performance of the multiple regression model using 5-fold cross-validation
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import KFold, cross_val_score

# Select predictors and dependent variable
predictors = ['GDP per capita', 'Social support', 'Healthy life expectancy', 'Freedom to make life choices', 'Generosity', 'Perceptions of corruption']
dependent_variable = 'Score'

# Prepare the data
X = data[predictors]
y = data[dependent_variable]

# Create a linear regression model
model = LinearRegression()

# Define the cross-validation method
cv = KFold(n_splits=5, shuffle=True, random_state=42)

# Calculate the cross-validated scores
scores = cross_val_score(model, X, y, cv=cv)

# Print the cross-validated scores
print("Cross-validated scores:", scores)
print("Mean cross-validated score:", np.mean(scores))
print("Standard deviation of cross-validated scores:", np.std(scores))
```

# **5-fold cross-validation**

In this analysis, we employ a linear regression model to predict happiness scores based on a range of factors, including GDP per capita, social support, healthy life expectancy, freedom to make life choices, generosity, and perceptions of corruption. To assess the model's performance and generalizability, we utilize 5-fold cross-validation. This approach partitions the dataset into five equal portions, and the model is trained iteratively on four parts while being tested on the remaining portion. This way, we can evaluate the model's capability to generalize to new, unseen data.

The cross-validation results provide R-squared values, which represent the proportion of the variance in the dependent variable that is predictable from the independent variables. The R-squared values range from approximately 0.60 to 0.86, with an average R-squared value of 0.748. This means that our linear regression model can, on average, explain about 74.8% of the variability in happiness scores using the given factors. However, the standard deviation of 0.0877 suggests some variability in the model's performance across different test sets, meaning the model may be more effective on some subsets of data than others.

Understanding the relationships between various factors and happiness scores, as indicated by the R-squared values, is valuable for policymakers and organizations. By identifying areas of focus to improve the well-being of their populations, they can make informed decisions that lead to better outcomes. Additionally, the R-squared values provide a way to compare the performance of different models, enabling the selection of the most suitable model for the given dataset. This, in turn, improves our comprehension of the factors that contribute to happiness and enhances the effectiveness of any interventions or initiatives aimed at increasing happiness in society.

```python
# Visualize the cross-validated scores to assess the model's performance across different subsets of the data
fig, ax = plt.subplots(figsize=(8, 6))
ax.bar(range(1, len(scores) + 1), scores)
ax.set_title('Cross-validated Scores')
ax.set_xlabel('Fold')
ax.set_ylabel('Score')
ax.set_xticks(range(1, len(scores) + 1))
plt.show()
```

In the first iteration (bar 1), the R-squared value is approximately 0.60, meaning that the model is able to explain 60% of the variability in happiness scores using the given factors. This is the lowest R-squared value among the five iterations. As we move to the second iteration (bar 2), the R-squared value improves to approximately 0.71, followed by further improvements in iterations 3, 4, and 5, where the R-squared values are approximately 0.78, 0.86, and 0.78, respectively. The fourth iteration (bar 4) exhibits the best performance, with the model explaining around 86% of the variability in happiness scores.

The variability in the R-squared values across the five iterations suggests that our linear regression model's performance is influenced by the specific data subsets used for training and testing. This highlights the importance of cross-validation in model evaluation, as it allows us to assess the model's generalizability and robustness to different data samples.

By understanding the model's performance across various subsets of data, we can better comprehend its strengths and limitations. The observed variability in R-squared values suggests that there may be room for improvement in the model, possibly by incorporating additional predictors or utilizing alternative modeling techniques. However, the average R-squared value of approximately 0.748 implies that our model is reasonably effective in explaining the variability in happiness scores, making it a valuable tool for understanding the factors that contribute to happiness in society.

```python
# Evaluate the model's performance on the new dataset using Mean Squared Error (MSE) and Mean Absolute Error (MAE)
from sklearn.metrics import mean_squared_error, mean_absolute_error

mse = mean_squared_error(data_2022['Score'], data_2022['Predicted Score'])
mae = mean_absolute_error(data_2022['Score'], data_2022['Predicted Score'])

print("Mean Squared Error (MSE):", mse)
print("Mean Absolute Error (MAE):", mae)
```

The evaluation metrics table, which includes Mean Squared Error (MSE) and Mean Absolute Error (MAE), helps us understand the accuracy of our linear regression model's predictions on the 2022 dataset. The MSE and MAE values indicate that the model's predictions have some degree of error, which means there is room for improvement.

The MSE, which is 0.275 in our case, takes the differences between the actual and predicted scores, squares them, and then averages the results. It's sensitive to large errors, so if the MSE value is high, it means our predictions have significant errors. On the other hand, the MAE, which is 0.412 for our model, is easier to interpret. It calculates the average of the absolute differences between actual and predicted scores, giving us a direct measure of the average error in our predictions.

So, looking at these values, we can see that our model is not perfect and could be improved to provide more accurate predictions of happiness scores.
